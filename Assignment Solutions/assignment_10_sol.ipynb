## 7. Prime Number Generator
#### Write a generator function genPrimes() that yields an infinite sequence of prime numbers starting from 2. 
#### Use a loop to print the first 5 prime numbers generated by genPrimes().

```python
def genPrime():
    start_prime_num = 2
    prime_num_lst = []
    while True:
        is_prime = True
        for prime in prime_num_lst:
            if start_prime_num%prime==0:
                is_prime = False
                break
        if is_prime:
            prime_num_lst.append(start_prime_num)
            yield start_prime_num
        start_prime_num+=1
get_nxt_prime = genPrime()
for _ in range(5):
    print(next(get_nxt_prime))
```
